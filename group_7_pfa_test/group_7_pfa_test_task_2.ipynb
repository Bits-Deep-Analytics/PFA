```python
## Please write your code here
## Only to be submitted by team leader

import pandas as pd
import numpy as np

#importing iris dataset from github url and reading it into pandas dataframe.
iris_dataset = pd.read_csv('https://raw.githubusercontent.com/Bits-Deep-Analytics/PFA/main/Iris.csv')
iris_dataset.head(10)
```

```python
#adding feature pl_pw ratio as PetalLengthCm/PetalWidthCm
iris_dataset['pl_pw'] = iris_dataset['PetalLengthCm']/iris_dataset['PetalWidthCm']
iris_dataset.head()
```

```python
#adding feature sl_pw ratio as SepalLengthCm/SepalWidthCm
iris_dataset['sl_pw'] = round(iris_dataset['SepalLengthCm']/iris_dataset['SepalWidthCm'], 2)
iris_dataset.head()
```

```python
#adding feature plw_slw ratio as pl_pw/sl_sw
iris_dataset['plw_slw'] = round(iris_dataset['pl_pw']/iris_dataset['sl_pw'], 2)
iris_dataset.head()
```

```python
#Removing all of the other raw predictors and only keep the derived attributes (pl_pw, sl_sw and plw_slw)
iris_dataset.drop(['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], axis=1, inplace = True)
```

```python
iris_dataset.to_csv('group_7_pfa_test_task-2.csv', index=False)
```
