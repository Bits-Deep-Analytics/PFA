```python
import pandas as pd

# Load the dataset
file_path = 'https://raw.githubusercontent.com/Bits-Deep-Analytics/PFA/main/Iris.csv'
iris_df = pd.read_csv(file_path)

# Create new features as per the requirements
iris_df['pl_pw'] = iris_df['PetalLengthCm'] / iris_df['PetalWidthCm']
iris_df['sl_sw'] = iris_df['SepalLengthCm'] / iris_df['SepalWidthCm']
iris_df['plw_slw'] = iris_df['pl_pw'] / iris_df['sl_sw']

# Remove all other raw predictors and keep only the derived attributes
derived_iris_df = iris_df[['pl_pw', 'sl_sw', 'plw_slw']]

# Display the first few rows of the modified dataframe
derived_iris_df.head()
```

```python
# Include the target variable 'Species' in the new dataset
derived_iris_with_species = derived_iris_df.assign(Species=iris_df['Species'])

# Define the file path for the new CSV file
new_file_path = '/content/group_12_pfa_test_task-2.csv'

# Save the new dataset as a CSV file
derived_iris_with_species.to_csv(new_file_path, index=False)
```
