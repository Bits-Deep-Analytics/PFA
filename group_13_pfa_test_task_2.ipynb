```python
import pandas as pd
import numpy as np
```

```python
url='https://raw.githubusercontent.com/Bits-Deep-Analytics/PFA/main/Iris.csv'
data=pd.read_csv(url)
data.head()
```

```python
data['pl_pw']=data['PetalLengthCm']/data['PetalWidthCm']
data.head()
```

```python
data['sl_sw']=data['SepalLengthCm']/data['SepalWidthCm']
data.head()
```

```python
data['plw_slw']=data['pl_pw']/data['sl_sw']
data.head()
```

```python
final_data=data.drop(columns=['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
final_data.head()
```

```python
final_data.to_csv('group_13_pfa_test_task-2.csv',index=False)
print('dataset saved successfully')
```
